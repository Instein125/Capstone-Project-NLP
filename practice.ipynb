Reading in Audio Files

```python
# Defining the paths to the clean audio and noise audio folders
clean_audio_folder = 'clean_audio'
noise_audio_folder = 'noise_audio'
trimmed_clean_folder = 'trimmed_clean_audio'
trimmed_noise_folder = 'trimmed_noise_audio'
output_folder = 'noised_speech'

# Ensure the output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Ensure the output folders exist
if not os.path.exists(trimmed_clean_folder):
    os.makedirs(trimmed_clean_folder)
if not os.path.exists(trimmed_noise_folder):
    os.makedirs(trimmed_noise_folder)
```

Trimming Audio

```python
# Function to trim audio and save
def trim_and_save(input_path, output_path):
    for filename in os.listdir(input_path):
        if filename.endswith(".wav"):
            file_path = os.path.join(input_path, filename)
            y, sr = librosa.load(file_path, sr=None)
            y_trimmed, _ = librosa.effects.trim(y, top_db=20)
            output_file_path = os.path.join(output_path, filename)
            sf.write(output_file_path, y_trimmed, sr)
            print(f"Trimmed audio saved at {output_file_path}")


# Trim and save clean audio
trim_and_save(clean_audio_folder, trimmed_clean_folder)

# Trim and save noise audio
trim_and_save(noise_audio_folder, trimmed_noise_folder)
```

# Spectogram

```python
noisy_speech, noisy_speech_sr = librosa.load(noise_speech[2], sr=None)
```

```python
S = librosa.feature.melspectrogram(y=noisy_speech,
                                   sr=noisy_speech_sr,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
S
```

```python
S.shape
```

```python
def plot_spect(y):
    # Plot the mel spectogram
    fig, ax = plt.subplots(figsize=(10, 5))
    img = librosa.display.specshow(S_db_mel,
                                x_axis='time',
                                y_axis='log',
                                ax=ax)
    ax.set_title('Mel Spectogram Example', fontsize=20)
    fig.colorbar(img, ax=ax, format=f'%0.2f')
    plt.show()
```

```python
plot_spect(S)
```

# Spectogram to audio

```python
A = librosa.feature.inverse.mel_to_audio(S)
```

```python
plot_waveform(A, "Contructed Audio")
```
