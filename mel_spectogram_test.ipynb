```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import os
import soundfile as sf

from glob import glob

import librosa
import librosa.display
import IPython.display as ipd

from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
HOP_LENGTH = 256
MONO = 22050
SAMPLE_RATE = 16000
FRAME_SIZE = 512
```

# Audio Player

```python
def play(signal):
    return ipd.Audio(signal)
```

# Loader

```python
def load(file_path):
        signal = librosa.load(file_path,
                              sr=SAMPLE_RATE,
                              mono=MONO)[0]
        return signal
```

# Spectogram Extractor

```python
def extractor(signal):
    mel_spec = librosa.feature.melspectrogram(y=signal,
                                   sr=SAMPLE_RATE,
                                   n_mels=128 * 2,)
    spectrogram = np.abs(mel_spec)
    mel_spectrogram = librosa.amplitude_to_db(spectrogram)
    return mel_spectrogram
```

# Spectogram to audio

```python
def convert_spectrograms_to_audio(spectrogram):
        # mel spectrogram -> spectrogram
        spec = librosa.db_to_amplitude(spectrogram)
        # apply Griffin-Lim
        signal = librosa.feature.inverse.mel_to_audio(spec, hop_length=HOP_LENGTH)
        return signal
```
